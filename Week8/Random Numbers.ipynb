```python
x=arange(0.,pi+.001,.01)
f=lambda x: sin(x)/2
F=lambda x: (cos(0)-cos(x))/2
plot(x,f(x),label="True distribution",lw=3)
plot(x,F(x),label="Cumulative distribution",lw=3)
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
def distribute1d(x,prob,N):
    """takes and numberical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data."""
    F = cumsum(prob)/(cumsum(prob)[N])
    g= searchsorted(F,z)
    xsamples=x[g]
    return xsamples
print x[g]
print distribute1d(x,f(x),N)
```
