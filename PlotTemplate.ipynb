(c) Add one cell to create the data arrays for the following functions over the range {5 <
y < 5}

E = $cosh(y)$

|p| = $sinh(y)$

Use regular python #comment formatting to describe the steps in this cell.

```python
import numpy as np #import numpy type arrays for use
import math #give access to math functions
step=.01#distance between each consecutive y
y=arange(-5.,5.+step,step)#y is every step from -5 to 5(y)
E=cosh(y) #define E
p=sinh(y) #supposed to be |p|, but p must be greater than or equal to 0 for this to be true.
```
