```python
from scipy.optimize import *
waveform_1=load('waveform_1.npy')
t=arange(len(waveform_1))
wf2=load('waveform_2.npy')
plot(t,waveform_1)
xlabel("Time (ns)")
ylabel("Intensity (bins)")
```

```python
def model(t, B, A, mu, sigma):
    return B+A*exp(-((t-mu)/sigma)**2.)
#mu = location, sigma = width
```

```python
x0=array([3.,30.,15.,1.])
popt,pcov=curve_fit(model,t,waveform_1,x0)
plot(t,waveform_1,label="data")
plot(t,model(t,popt[0],popt[1],popt[2],popt[3]),label="fit",linewidth=2)
legend(loc="best")
show()
```

```python
def model3(t, A, B, C, D, m1, s1,m2,s2,m3,s3):
    return A+B*exp(-((t-m1)/s1)**2.)+C*exp(-((t-m2)/s2)**2.)+D*exp(-((t-m3)/s3)**2.)
```

```python
x=array([3.,30.,12.,8.,15.,1.,23.,1.,28.,1.5])
popt,pcov=curve_fit(model3,t,wf2,x)
mod=model3(t,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5],popt[6],popt[7],popt[8],popt[9])
u=abs(mod-wf2)
print u
```
