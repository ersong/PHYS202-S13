```python
import numpy as np
step=.1
th=arange(-pi,pi+step/2,step)
y=(cos(2*th)*cos(th/2.))**2
#print len(th),len(y)
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plot(th,y,color="red",linewidth=2.5)
plot(th,y,'r*',color="blue")
xlabel("intensity", fontsize=15)
ylabel(r"$\theta$ (rad)",fontsize=15)
xticks(arange(-pi,pi+.3,2*pi/4),('$-\pi$',r'$-\frac{\pi}{2}$','0',r'$+\frac{\pi}{2}$','$+\pi$'),fontsize=15)
show()
```

```python
import numpy as np
import matplotlib
import math
dx=.001
t=arange(-5,5+dx/2,dx)
f=(t+1)/(t-1)
g=e**(-t)
plot(t,f,color="blue",label="(1+t)/(t-1)")
plot(t,g,color="red",linestyle=":",label="exp(-t)")
xlim(-5,5)
ylim(-10,10)
xlabel("time (seconds)", fontsize=15)
ylabel("f(t),g(t)",fontsize=15)
legend(loc='lower left')
show()
```

```python
import scipy.integrate
def pointField(x,y,q,Xq,Yq):
    """Takes array (x,y), charge q, and position Xq,Yq, and returns tuple (Ex,Ey)"""
    k=8.98e9
    Ex=k*q*(x-Xq)/((x-Xq)**2+(y-Yq)**2)**1.5
    Ey=k*q*(y-Yq)/((x-Xq)**2+(y-Yq)**2)**1.5
    return (Ex,Ey)
a=1
b=2
lam=1e-6
dx=.00001
x=arange(a,b+dx/2,dx)
y=x
E1,E2=pointField(x,y,lam*dx,3,3)
I1=trapz(E1)
I2=trapz(E2)
print I1,I2
```
