Electric Dipole Equation
--------------------------

$V(x,y)=\dfrac{kq}{\sqrt{x^2+(y-d)^2}}-\dfrac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
dx=.06
dy=.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
k=1;q=1;d=1;c=0
V=k*q/sqrt(X**2.+(Y-d)**2.) - k*q/sqrt(X**2.+(Y+d)**2.)
U=k*q/sqrt((X-c)**2.+Y**2.) - k*q/sqrt((X+c)**2.+Y**2.)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
#help(Axes3d.plot_surface)
cb=fig.colorbar(p,shrink=.5)
```

```python
#contours
levels=array([8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8])

contour(X,Y,V,levels,linewidths=4,cmap=cm.autumn)
```

```python
k=8.98#N*m^2/C^2
q=1e-9#C
d=2
c=2
dx=12/101.
dy=12/101.
x=arange(-6,6,dx)
y=arange(-6,6,dy)
X,Y=meshgrid(x,y)
V=k*q/sqrt(X**2.+(Y-d)**2.) + k*q/sqrt(X**2.+(Y+d)**2.)
U=0-k*q/sqrt((X-c)**2.+Y**2.) - k*q/sqrt((X+c)**2.+Y**2.)
levels=array([-1e-7,-1e-8,-1e-9,-1e-10,0,1e-10,1e-9,1e-8,1e-7])
contour(X,Y,U+V,levels,linewidths=4,cmap=cm.jet)
```

```python
def pointPotential(x,y,q,posx,posy):
    """Solves for V when given center point (x,y) and location (posx, posy)"""
    Vxy=(8.89e9)*q/sqrt((X-posx)**2.+(Y-posy)**2.)
    return Vxy
```

```python
def dipolePotential(x,y,q,d):
    """k=Coulomb's constant, 8.9e9. Set Vxy with the equation below to check total potential in the x axis at posx"""
    k=8.9e-9
    Vxy = k*q/sqrt((X-posx)**2.+Y**2.) + k*q/sqrt((X+posx)**2.+Y**2.)
    return Vxy
```
