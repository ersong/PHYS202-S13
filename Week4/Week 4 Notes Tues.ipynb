Electric Dipole Equation
--------------------------

$V(x,y)=\dfrac{kq}{\sqrt{x^2+(y-d)^2}}-\dfrac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
dx=.06
dy=.06
x=arange(-2.,2.,dx)
y=arange(-2.,2.,dy)
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
k=1;q=1;d=1;c=0
V=k*q/sqrt(X**2.+(Y-d)**2.) - k*q/sqrt(X**2.+(Y+d)**2.)
U=k*q/sqrt((X-c)**2.+Y**2.) - k*q/sqrt((X+c)**2.+Y**2.)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(X,Y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
#help(Axes3d.plot_surface)
cb=fig.colorbar(p,shrink=.5)
```

```python
#contours
levels=array([8,4,2,1,.5,.25,0,-.25,-.5,-1,-2,-4,-8])

contour(X,Y,V,levels,linewidths=4,cmap=cm.autumn)
```

```python
def pointPotential(x,y,q,posx,posy):
    """Give charge q in Coulombs, charge location (x,y) in meters, and point of interest (posx, posy) in meters"""
    Vxy=(8.8975517873681764e9)*q/sqrt((x-posx)**2.+(y-posy)**2.)
    return Vxy

x2 = 1.; y2 = 1.
V11 = pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
def dipolePotential(x,y,q,d,angle):
    """Give charge q in Coulombs, charge location (x,y) in meters, distance d in meters, and angle of rotation, angle"""
    k=8.8975517873681764e9
    a=(d/2.)*cos(angle)
    b=(d/2.)*sin(angle)
    c1=sqrt((x-a)**2.+(y-b)**2.)
    c2=sqrt((x+a)**2.+(y+b)**2.)
    kq=k*q
    Vxy = kq/c1-kq/c2
    return Vxy

d=1
angle=pi/3
q=1e-9
x,y = meshgrid(arange(-2.,2.,0.03),arange(-2.,2.,0.03))
Vxy = dipolePotential(x,y,q,d,angle)
print Vxy
```

```python
dx=.04
dy=.04
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dx))
q=1e-9
d=1
angle=pi/4
Vxy=abs(dipolePotential(x,y,q,d,angle))
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(x,y,Vxy,rstride=1,cstride=1,linewidth=0,cmap=cm.BuGn)
cb=fig.colorbar(p,shrink=.5)
```
