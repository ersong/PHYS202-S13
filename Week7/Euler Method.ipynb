```python
# v(n+1) = v(n) + a(n)dt
# r(n+1) = r(n) + v(n+1)dt
import numpy as np
dx1=.25
dx2=.02
#print len(x),len(y)
def solve_y_euler(x0,y0,dx,xf):
    x=np.arange(x0,xf+dx/2,dx)
    y=zeros(shape(x))
    for i in range(len(x)):
        y[i]=(x[i]+2*y0)*dx+y0
        y0=y[i]
    return y
ya= solve_y_euler(0,0,dx1,1)
yb= solve_y_euler(0,0,dx2,1)
print ya[len(ya)-1]
print yb[len(yb)-1]
```
