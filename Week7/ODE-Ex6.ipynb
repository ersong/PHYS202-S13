a) Euler's Method

```python
# v(n+1) = v(n) + a(n)dt
# r(n+1) = r(n) + v(n+1)dt
import numpy as np
dx1=.25
dx2=.02
#print len(x),len(y)
def solve_y_euler(x0,y0,dx,xf):
    x=np.arange(x0,xf+dx/2,dx)
    y=zeros(shape(x))
    for i in range(len(x)):
        y[i]=(x[i]+2*y0)*dx+y0
        y0=y[i]
    return y
ya= solve_y_euler(0,0,dx1,1)
yb= solve_y_euler(0,0,dx2,1)
print ya[len(ya)-1]
print yb[len(yb)-1]
```

```python
y=lambda x: .25*e**(2*x)-.5*x-.25
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
N=1000.
xf=1.
dx=float(1/N)
x=arange(0,xf+dx/2,dx)
x1=arange(0,xf+dx/2,dx1)
x2=arange(0,xf+dx/2,dx2)
y1= solve_y_euler(0,0,dx1,xf)
y2= solve_y_euler(0,0,dx2,xf)
plot(x,y(x),label="Exact Solution")
plot(x1,y1,label="dx=.25")
plot(x2,y2,label="dx=.02")
legend(loc='upper left')
print "Step of ",dx2," gives most accurate result."
```

c) Euler Cromer + Midpoint

```python
def solve_y_ec(x0,y0,dx,xf):
    x=np.arange(x0,xf+dx/2,dx)
    y=zeros(shape(x))
    z=zeros(shape(x))
    for i in range(len(x)):
        y[i]=(x[i]+2*y0)*dx+y0
        z[i]=(x[i]+2*y[i])*dx+y0
        y0=y[i]
    return z
def solve_y_mp(x0,y0,dx,xf):
    x=np.arange(x0,xf+dx/2,dx)
    y=zeros(shape(x))
    z=zeros(shape(x))
    for i in range(len(x)):
        y[i]=(x[i]+2*y0)*dx+y0
        z[i]=.5*((x[i]+2*y0)+(x[i]+2*y[i]))*dx+y0
        y0=y[i]
    return z
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
dx1=.25
dx2=.02
N=10000.
xf=1.
dx=float(1/N)
x=arange(0,xf+dx/2,dx)
x1=arange(0,xf+dx/2,dx1)
x2=arange(0,xf+dx/2,dx2)
y1= solve_y_mp(0,0,dx1,xf)
y2= solve_y_mp(0,0,dx2,xf)
y3= solve_y_ec(0,0,dx1,xf)
y4= solve_y_ec(0,0,dx2,xf)
ya= solve_y_euler(0,0,dx1,1)
yb= solve_y_euler(0,0,dx2,1)
plot(x1,y(x1)-y3,label="Exact - Euler Cromer, step=.25")
plot(x2,y(x2)-y4,label="Exact - Euler Cromer, step=.02")
#plot(x1,y(x1)-ya,label="Exact - Euler, step=.25")
#plot(x2,y(x2)-yb,label="Exact - Euler, step=.02")
plot(x1,y(x1)-y1,label="Exact - Midpoint, step=.25")
plot(x2,y(x2)-y2,label="Exact - Midpoint, step=.02")
legend(loc='lower left')
show()
```
