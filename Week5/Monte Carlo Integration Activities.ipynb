Hit or Miss MC Integration
--------------------------

1.   Define the function we want to integrate (it could also be an array of points).

II.  Create the bounding box around our function within whih to work

III. Randomy (or selectively randomly) sample points within that region

IV.  Calculate the fraction of points that fall within the function boundary.

V.   The integral is then the point fraction times the area of the bounding box.

Example 1
=========
We'll start with a simple 1 dimensional example and implement the most basic hit or miss method using truly random samples within our domain. Let's try to write the code for the series of steps described above.


Here is the function we'll work with:

$f(x)=x^{1/2}+sin(x)$

Let's define it and plot it so we have an idea of what we are looking at (Step 1).

```python
f=lambda x: x**(.5)+sin(x)
x=arange(0.,15.,.001)
myplot=plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
xlabel(r'$x$',fontsize=20)
xlim(0.,15.)
show()
```

```python
xmin=1.7
xmax=12.4
subx=x[logical_and((x>xmin),(x<xmax))]
ymax=max(f(subx))*1.05
ymin=0.
myplot=plot(x,f(x),lw=3)
xlim=(0.,15.)
plot([xmin,xmin],[ymin,ymax],color='k',linestyle='--')
plot([xmax,xmax],[ymin,ymax],color='k',linestyle='--')
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--')
show()
```

```python
N=10000
samples_x=xmin+(xmax-xmin)*random_sample(N);
samples_y=ymin+(ymax-ymin)*random_sample(N);
myplot=plot(x,f(x),lw=3)
xlim=(0.,15.)
plot(samples_x[1:200],samples_y[1:200],'.')
plot([xmin,xmin],[ymin,ymax],color='k',linestyle='--')
plot([xmax,xmax],[ymin,ymax],color='k',linestyle='--')
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--')
show()
```

```python
newmask=(samples_y<f(samples_x))
sum(newmask)
```

```python
myplot=plot(x,f(x),lw=3)
xlim=(0.,15.)
plot(samples_x[1:200],samples_y[1:200],'.')
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')
plot([xmin,xmin],[ymin,ymax],color='k',linestyle='--')
plot([xmax,xmax],[ymin,ymax],color='k',linestyle='--')
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--')
show()
```

```python
I=(xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

```python
Iana=lambda x: (2/3.)*x**(1.5)-cos(x)
Iexact=Iana(xmax)-Iana(xmin)
Ierr=100*abs(Iexact-I)/Iexact
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```
