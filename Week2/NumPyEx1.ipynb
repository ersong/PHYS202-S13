```python
def eratosthenes_sieve(maximum): # multiples of a prime number except the prime number aren't prime
    numlist=range(maximum+1)
    for n in numlist:
        for a in numlist: #multiples of 2 except 2 aren't prime
            if a>2 and a%2==0:
                numlist.remove(a)
        for b in numlist:  #multiples of 3 except 3 aren't prime
            if b>3 and b%3==0:
                numlist.remove(b)
        for c in numlist: #multiples of 5 except 5 aren't prime
            if c>5 and c%5==0:
                numlist.remove(c)
        for d in numlist: #multiples of 7 except 7 aren't prime
            if d%7==0:
                numlist.remove(d)
        return numlist

print eratosthenes_sieve(30)#works
print eratosthenes_sieve(10)
print eratosthenes_sieve(100)
```
