Reproduce the following graph of these two functions for the four values of k shown. Let

(c) Finally, add code to calculate the running average of the data. Plot both the original
data and the running average on the same graph, again over the range covered by the
rst 1000 data points. Be sure to include a legend and make it "pretty" (shorthand for
publication quality).

```python
data = loadtxt("sunspots.txt",float)
a = data[0:1000,0]
b = data[0:1000,1]
figure(num=None, figsize=(10, 6), dpi=80, facecolor='w', edgecolor='k')
plot(a,b,color="orange",linewidth=2.5,label="Sunspot Number")
xlabel("Year", fontsize=20)
ylabel("Sunspots",fontsize=20)
dates=[]
for i in arange(0,1100/12.,1000./(10*12)):
    dates.append(int(float(i)+1749))

total=0
avg=[]
spot=0
for s in b:
    spot+=1
    total+=s
    avg.append(float(total/spot))
plot(a,avg,color="blue",linewidth=2.5,label="Running Average")
xticks(arange(0,1100,1000./10),(dates),fontsize=15)
title("1000 Month Sunspot Running Average", fontsize=20)
legend()
grid(True)
show()
```
